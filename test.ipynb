## Data Cleaning

```python
#Selecting relevant columns.
df1 = df[['communities_villages', 'water_point_id', 'water_functioning', 'water_point_condition', 'water_source_type', 'water_point_geocode']].copy()

df1.head()
```

```python
df1.sample(15)
```

```python
#Checking counts of unique values for each column.
for col in df1.columns:
    print(col, df1[col].nunique(), len(df1))
```

```python
#Checking for null values.
df1.isnull().sum()
```

```python
#Viewing missing data
df1[df1['water_point_geocode'].isnull()]
```

```python
#Replacing the missing water_point_geocode with their unique water_point_ids
df1['water_point_geocode'].fillna(df1.water_point_id, inplace=True)

df1['water_point_geocode'].isnull().sum()
```

```python
#Checking for duplicates
df1.duplicated().sum()
```

```python
df1[df1.duplicated()]
```

```python
len(df1)
```

```python
#Dropping the duplicate.
df1.drop_duplicates(inplace=True)

print(len(df1))
type(df1)
```

## Data Analysis

**How many water points are functional?**

```python
df1.water_functioning.value_counts()
```

**ANS = There are 622 functional water points.**

**Calculating the number of water points per community:**

```python
#Grouping the communities and counting the unique water points in each using water_point_geocode.
pd.set_option('display.max_rows', None)
number_water_points = df1.groupby('communities_villages')['water_point_geocode'].count().reset_index(name = 'waterpoints_count')

print(number_water_points)
```

```python
number_water_points['waterpoints_count'].sum()
```

**What is the rank for each community by the percentage of broken water points?**

```python
#Unique values of water_point condition.
df1.water_point_condition.value_counts()
```

```python
#Grouping the dataframe by communities_villages and water_point_condition to find the counts of each condition.
df2 = df1.groupby(['communities_villages', 'water_point_condition'])['water_point_geocode'].count().reset_index(name='count')

df2.head(15)
```

```python
#Reorganizing the dataframe into a more useful format using pivoting.
pivoted = df2.pivot_table(index='communities_villages', columns='water_point_condition', values='count', fill_value=0).reset_index()

pivoted.head()
```

```python
#Merging the two dataframes: number_water_points and pivoted in order to calculate rank.
new_df = pd.merge(number_water_points, pivoted)

new_df.head(10)
```

```python
#Calculating the community ranks, correct to 2 decimal places.
new_df['community_rank'] = new_df.apply(lambda row: (row['broken'] / row['waterpoints_count'])*100, axis=1).round(2)

new_df.head()
```

**ANS:**

```python
#Sorting and displaying final results.
pd.set_option('display.max_rows', None)
result = new_df[['communities_villages', 'community_rank']].sort_values(by=['community_rank'])

print(result)
```
